```python

def is_positive(sentiment):
    if sentiment == 'positive':
        return 1
    else:
        return 0


def is_negative(sentiment):
    if sentiment == 'negative':
        return 1
    else:
        return 0

reviews['Positive'] = reviews['Sentiment'].apply(is_positive)
reviews['Negative'] = reviews['Sentiment'].apply(is_negative)


```

```python
result=reviews.groupby('Year')[['Positive', 'Negative']].sum()
```

```python
result.plot(kind='bar', figsize=(10, 6))
plt.title('Trend of Positive and Negative Sentiment Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.legend()
plt.show()
```

```python

import numpy as np

sum_positive_annual = reviews.groupby('Year')['Positive'].sum()


years = sum_positive_annual.index
matrix = pd.DataFrame(index=years, columns=years, dtype=float)


for row_year in years:
    for col_year in years:
        if col_year < row_year:
      
            previous_year_sum = sum_positive_annual.loc[col_year]
            current_year_sum = sum_positive_annual.loc[row_year]
            percentage_change = ((current_year_sum - previous_year_sum) / previous_year_sum) * 100
            matrix.at[row_year, col_year] = round(percentage_change, 2)
        else:
     
            matrix.at[row_year, col_year] = 0

styled_matrix = matrix.style.background_gradient(cmap='Greens')

styled_matrix
```

```python
import numpy as np

reviews_from_2000 = reviews[reviews['Year'] >= 2000]
sum_negative_annual = reviews_from_2000.groupby('Year')['Negative'].sum()


years = sum_negative_annual.index
matrix = pd.DataFrame(index=years, columns=years, dtype=float)


for row_year in years:
    for col_year in years:
        if col_year < row_year:
          
            previous_year_sum = sum_negative_annual.loc[col_year]
            current_year_sum = sum_negative_annual.loc[row_year]
            percentage_change = ((current_year_sum - previous_year_sum) / previous_year_sum) * 100
            matrix.at[row_year, col_year] = round(percentage_change, 2)
        else:
          
            matrix.at[row_year, col_year] = 0

styled_matrix = matrix.style.background_gradient(cmap='Blues')
styled_matrix
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Download the necessary NLTK data (if you haven't already)
nltk.download('punkt')

# Applying NLTK's tokenizer
reviews['Token'] = reviews['CleanedText'].apply(lambda x: word_tokenize(x))

# The 'tokens' column now contains more accurately tokenized words
```

```python
reviews
```

```python
from gensim.corpora import Dictionary
from gensim.models import LdaModel

# Convert the 'tokens' column into a list of lists for Gensim
documents = reviews['Token'].tolist()

# Create a Gensim Dictionary
dictionary = Dictionary(documents)

# Filter extremes to refine the dictionary
dictionary.filter_extremes(no_below=5, no_above=0.5)

# Create a Corpus
corpus = [dictionary.doc2bow(doc) for doc in documents]

# Train the LDA model
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=2, passes=10)

# Explore the topics
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic: {idx} \nWords: {topic}")
```

```python
def assign_dominant_topic(ldamodel, corpus):
    dominant_topics = []
    for _, row_list in enumerate(ldamodel[corpus]):
        row = sorted(row_list, key=lambda x: (x[1]), reverse=True)
        dominant_topic_num = row[0][0] if row else None
        dominant_topics.append(dominant_topic_num)
    return dominant_topics

# Use the trained model and the corpus you already created
reviews['dominant_topic'] = assign_dominant_topic(lda_model, corpus)
```
